```python
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df["pubdate"] = df["pubdate"].apply(lambda x: datetime.datetime.fromtimestamp(x))
```

```python
plt.figure(figsize=(12, 6))

df["pubdate"].dt.to_period("d").value_counts().sort_index().plot(kind="bar")

plt.title('Распределение публикаций по дням')
plt.xlabel('День')
plt.ylabel('Количество публикаций')
plt.xticks(rotation=45)
plt.show()
```

```python
df['title_length'] = df['title'].apply(lambda x: len(str(x)))
df['snippet_length'] = df['snippet'].apply(lambda x: len(str(x)))

plt.figure(figsize=(12, 6))
sns.histplot(df['title_length'], bins=30)

plt.title('Распределение длины заголовков')
plt.xlabel('Длина заголовка')
plt.ylabel('Частота')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df['snippet_length'], bins=30)
plt.title('Распределение длины сниппетов')
plt.xlabel('Длина сниппета')
plt.ylabel('Частота')
plt.show()
```
